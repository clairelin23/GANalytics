## This notebook is for visualizing results

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Define the directory containing the PNG images
image_dir = 'plotting_pokemon'

# Load the PNG images into a list
images = []
for filename in os.listdir(image_dir):
    if filename.endswith('.png'):
        img_path = os.path.join(image_dir, filename)
        img = Image.open(img_path)
        images.append(img)

# Create a 6x6 grid of subplots and plot the images
fig, axs = plt.subplots(nrows=5, ncols=5, figsize=(7, 7))
axs = axs.flatten()

for i, img in enumerate(images):
    axs[i].imshow(img)
    axs[i].axis('off')

# Adjust the spacing between the subplots
plt.subplots_adjust(wspace=0.01, hspace=0.05)

plt.tight_layout()
plt.show()
```
